 <h1>Web Scrapping On Insta Hyre Website: </h1>
 <h5>Web scraping to extract 1500 job listings related to Python roles from the Instahyre website.Accompanied with the dataset that comprises of specific details for each job listing.</h5>

<h3>WebDriver Initialization and Basic Commands:</h3>

```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time as t
```

<h4>Opening the Instahyre webpage:</h4>

```python
url="https://www.instahyre.com/python-jobs"                         
driver=webdriver.Chrome()
driver.get(url)
```

```python
n=[]
lo=[]
f=[]                         #Creating empty arrays
e=[]        
a=[]
s=[]
li=[]

# Extracting data using xpath for 1500 jo listings

for j in range(75):
    name=driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']/a/div/div[@class='company-name ng-binding']")
    loc=driver.find_elements("xpath","//div[@class='employer-locations']/span/span[@class='ng-binding ng-scope']")
    found=driver.find_elements("xpath","//div[@class='employer-info']/span[@ng-if='opp.employer.company_founded']")
    empl=driver.find_elements("xpath","//div[@class='employer-info']/span[@ng-if='opp.employer.employee_count']")
    about=driver.find_elements("xpath","//div[@ng-if='opp.employer.instahyre_note']")
    skill=driver.find_elements("xpath","//div[@ng-if='opp.job.keywords']/ul")
    link=driver.find_elements("xpath","//div[@class='opportunity-action-links opportunity-action-links-desktop col-md-2 col-sm-2 col-xs-12 pull-right']/div/a")

    # In single page we have 20 job listings,so applying loop in range 20
    
    for i in range(20):
        
        try:
            n.append(name[i].text)
        except Exception:
            n.append("No_data")  
            
        try:
            lo.append(loc[i].text)
        except Exception:
            lo.append("No_data")    
             
        try:
            f.append(found[i].text)
        except Exception:
            f.append("No_data") 
            
        try:
            e.append(empl[i].text)
        except Exception:
            e.append("No_data")   
        
        try:
            a.append(about[i].text)
        except Exception:
            a.append("No_data") 
        
        try:
            s.append(skill[i].text)
        except Exception:
            s.append("No_data") 
            
        try:
            li.append(link[i].get_attribute("href"))
        except Exception:
            li.append("No_data")
     
    #  To navigate to the Next page ,using click function
     
    next_button=driver.find_element("xpath","//div[@class='pagination ng-scope']/li[@ng-click='nextPage()']")
    next_button.click()

# Creating a dictionary named "dict",with following key:value pair 
      
dict={"job_position":n,"job_location":lo,"establishment_year":f,"employees":e,"About":a,"Skills":s,"Link":li}
df=pd.DataFrame(dict)                                      #Now,creating dataframe named "df" using pandas 
print(df)
df.to_csv("insta_hire.csv",index=False)                    #Creating csv file named "insta_hire.csv"
```
